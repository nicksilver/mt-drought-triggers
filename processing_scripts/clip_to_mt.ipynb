## Clip drought monitor data to Montana
This notebook takes downloaded drought monitor data and clips it to the state of Montana. 

```python
import geopandas as gpd
import glob
import numpy as np
import shutil
```

```python
dpath =  "/media/nick/Seagate Backup Plus Drive/Data/Drought_Monitor/"
mtpath =  "/media/nick/Seagate Backup Plus Drive/Data/Montana_GIS/"
fpaths = glob.glob(dpath + "*.zip")
mt = gpd.read_file("zip://" + mtpath + "MontanaStateBoundary.zip")
```

```python
fpaths[420]
```

```python
write_path =  dpath + 'Montana_Clipped/'
for f in fpaths:   
    fname = f.split("/")[-1].split(".")[0] + "_MT"
    data = gpd.read_file("zip://" + f)     
    data = data.to_crs(epsg=32100)
    data_clipped = gpd.clip(data, mt)
    if not data_clipped.empty:
        data_clipped.to_file(filename=write_path + fname, driver='ESRI Shapefile')
        print("File " + fname + " has been created.")
    else:
        print("File " + fname + " was unable to be processed. Saving empty MT shapefile as placeholder for date.")
        mt['DM'] = np.repeat(0, len(mt))
        mt.to_file(filename=write_path + fname, driver='ESRI Shapefile')
```

```python
mt_dirs = glob.glob(write_path + '*_M_MT')
for dir in mt_dirs:
    if dir.split('.')[-1] == 'zip':
        continue 
    shutil.make_archive(dir, 'zip', dir)
    shutil.rmtree(dir)
```
